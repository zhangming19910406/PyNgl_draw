```python
from __future__ import print_function
import os
import numpy as np
import numpy.ma as ma
import Ngl, Nio
from datetime import date, timedelta
```

```python
def draw_by_month(time, lat, lon, wave):
    # time is int32
    
    # calculate file name
    days = time/24 
    start = date(1900,1,1)
    delta = timedelta(days)
    offset = start + delta
    name = offset.isoformat()[:7]
    print(name)
    
    wks_type = "png"
    wks = Ngl.open_wks(wks_type, name)
    cnres = Ngl.Resources()

    # Contour resources
    cnres.cnFillOn        = True
    cnres.cnFillPalette   = "WhiteBlueGreenYellowRed"      # New in PyNGL 1.5.0
    cnres.cnLinesOn       = False
    cnres.cnLineLabelsOn  = False
    cnres.cnLevelSpacingF        = 0.5
    cnres.cnLevelSelectionMode = "ManualLevels"
    cnres.cnMinLevelValF         = 0.5
    cnres.cnMaxLevelValF         = 5.5

    # Labelbar resource
    cnres.lbOrientation   = "horizontal"
    cnres.lbLabelFontHeightF  = 0.010

    # title
    cnres.tiMainString    = name

    # Scalar field resources
    cnres.sfXArray        = lon
    cnres.sfYArray        = lat
    
    plot = Ngl.contour_map(wks,wave,cnres)
    
#     txres                = Ngl.Resources()         # text mods desired
#     txres.txFont         = 50
#     txres.txFontHeightF  = 0.02                    # font smaller. default big
#     txres.txFontOpacityF = 0.10                    # highly transparent
#     txres.txFontOpacityF = 0.5                    # half transparent
#     txres.txAngleF       = 35.
    
#     txid = Ngl.add_text(wks,plot,"Copyright reseved by ZhangMing",10,15,txres) 
    
    Ngl.draw(plot)        # Drawing plot will draw attached text
#     Ngl.frame(wks)        # Advance frame
    
    Ngl.delete_wks(wks)
```

## use netCDF

```python
from netCDF4 import Dataset
nc_obj=Dataset('./wave2.nc')
```

```python
print(nc_obj)
print(nc_obj.variables.keys())
for i in nc_obj.variables.keys():
    print(i)
```

```python
wave1=(nc_obj.variables['p0001'][:])
# wave2=(nc_obj.variables['p0005'][:])
longitude = nc_obj.variables['longitude'][:]
latitude = nc_obj.variables['latitude'][:]
```

```python
times = nc_obj.variables['time'][:]
```

```python
wave_index = 0
for time in times:
    wave_to_draw  = np.mean(wave1[wave_index,:,:,:], axis=0)
    draw_by_month(time, latitude, longitude, wave_to_draw)
    wave_index = wave_index +1
```
